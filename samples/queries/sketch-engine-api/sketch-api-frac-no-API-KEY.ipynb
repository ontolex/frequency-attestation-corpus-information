```python
# os,sys,glob
import os
import os.path
from os import path
import sys
import glob 
import locale
import string

# rdf
import rdflib
from rdflib import Graph
from rdflib.namespace import RDF, RDFS, XSD, OWL, DCAT, FOAF, SKOS
from rdflib import URIRef, BNode, Literal, Bag, Seq
from rdflib.tools.rdf2dot import rdf2dot

ITSRDF=rdflib.Namespace("http://www.w3.org/2005/11/its/rdf#")
DC = rdflib.Namespace("http://purl.org/dc/elements/1.1/") 
DCT = rdflib.Namespace("http://purl.org/dc/terms/") 
MS = rdflib.Namespace("http://w3id.org/meta-share/meta-share/") 
OLIA = rdflib.Namespace("http://purl.org/olia/discourse/olia_discourse.owl#")
LIME = rdflib.Namespace("http://www.w3.org/ns/lemon/lime")
LEXINFO =  rdflib.Namespace("http://www.lexinfo.net/ontology/2.0/lexinfo#")
ONTOLEX  = rdflib.Namespace("http://www.w3.org/ns/lemon/ontolex#")
FRAC = rdflib.Namespace("http://www.w3.org/ns/lemon/frac#")
DEMO = URIRef("http://llod.jerteh.rs/OntoLex-FrAC#")

```

```python
# manage reading words and generate graph triples
class Lexicon:    

    def __init__(self):
        self.lemma ="risk"
        self.pos= "NOUN"
        self.lng = "en"
        self.pos1= "n"
        self.le_url = "le_"+self.lemma+"_"+self.pos1
        self.cql = "[lemma ='risk' & tag='N.*'']"
        # collocations
        self.coll="""requests.get(base_url + '/collx', auth=(USERNAME, API_KEY), params={
        'q':'q[lemma="risk" & tag="N.*"]','corpname': 'preloaded/bnc2',}).json()"""
        self.colli="""https://app.sketchengine.eu/#concordance?corpname=preloaded/bnc2_tt31&tab=advanced&
        queryselector=cql&attrs=word,tag&viewmode=kwic&attr_allpos=all&refs_up=0&shorten_refs=1&glue=1&
        gdexcnt=300&show_gdex_scores=0&itemsPerPage=20&structs=s,g&refs==bncdoc.alltyp&default_attr=lemma&
        cql=[lemma="risk" & tag="N.*"]&showresults=1&results_screen=collocations&showTBL=0&tbl_template=&
        gdexconf=&f_tab=basic&f_showrelfrq=1&f_showperc=0&f_showreldens=0&f_showreltt=0&c_customrange=0&
        operations=[{"name":"cql","arg":"[lemma=\"risk\" & tag=\"N.*\"]",
        "query":{"queryselector":"cqlrow","cql":"[lemma=\"risk\" & tag=\"N.*\"]","default_attr":"lemma"},"id":6424}]"""
        # thesaurus
        self.thes ="""requests.get(base_url+'/thes', params={'corpname': 'preloaded/bnc2', 
        'lemma': 'risk', 'lpos': '-n',}, auth=(USERNAME, API_KEY)).json()"""
        self.thesi="""https://app.sketchengine.eu/#thesaurus?corpname=preloaded/bnc2_tt31&tab=advanced&
        lemma=risk&lpos=-n&showScores=1&showresults=1&itemsPerPage=10"""
        # wordsketch
        self.wsketch = """requests.get(base_url + '/wsketch', params= {'corpname': 'preloaded/bnc2',
        'lemma': 'make','lpos': '-n',}, auth=(USERNAME, API_KEY)).json() """
        self.wsketchi = """https://app.sketchengine.eu/#thesaurus?corpname=preloaded/bnc2_tt31&tab=advanced&
        lemma=risk&lpos=-n&showScores=1&showresults=1&itemsPerPage=10"""
        # documentation?
        self.cql_lng_url="https://www.sketchengine.eu/apidoc/"
        # or this https://www.sketchengine.eu/documentation/corpus-querying/"
        
        self.file_name = "demoFracQuery.ttl"
        self.url="http://llod.jerteh.rs/OntoLex-FrAC#" 
        return
        
    def getStats(self,data,target_n):
    # data=[{'s': '14.87162', 'n': 't'}, {'s': '9.05346', 'n': 'm'}, {'s': '8.78827', 'n': 'd'}],
        for item in data:
          if item['n'] == target_n:
            return item['s']   
        return None

    # graph initialisation for one novel 
    def  init_glexicon(self,g):      
       # for graph
       glex = URIRef(self.url+"QLEX")  
       
       #:SMD dct:creator " Ranka Stanković ....."@en;
       #dct:description "Test Lexicon for LLOD"@en;       
       #lime:lexicalEntries "5"^^<xsd:integer>;       
       #ontolex:entry :le_risk_8754
    
       g.add( (glex, RDF.type, LIME.Lexicon  ) ) 
       g.add( (glex, LIME.linguisticCatalog, URIRef("http://www.lexinfo.net/ontologies/2.0/lexinfo")) )
       g.add( (glex, DC.Language, Literal(self.lng)))
       g.add( (glex, MS.Language, Literal(self.lng)))
       
       # :BNC a owl:Class;
       #  rdfs:subClassOf [a owl:Restriction ; 
       #                   owl:onProperty frac:corpus ; 
       #                   owl:hasValue <https://app.sketchengine.eu/#dashboard?corpname=preloaded/bnc2_tt31>]. 
             
       gcor = URIRef(self.url+"BNC")
       g.add( (gcor, RDF.type, OWL.Class))
       bn = BNode()
       g.add( (bn, RDF.type, OWL.Restriction))
       g.add( (bn, OWL.onProperty, FRAC.corpus))
       g.add( (bn, OWL.hasValue, URIRef("https://app.sketchengine.eu/#dashboard?corpname=preloaded/bnc2_tt31")) )
       g.add( (gcor, RDFS.subClassOf, bn))
             
 
        
       #:q_thes_le_risk_n a frac:QueryResult;
       #    rdf:value """[lemma ="risk" & tag="N.*"]"""^^xsd:string;
       #    rdfs:annotation """https://app.sketchengine.eu/#thesaurus?corpname=preloaded/bnc2_tt31&tab=advanced&lemma=risk&lpos=-n&showScores=1&showresults=1&itemsPerPage=10""";
       #    owl:sameAs <https://noske.rgf.rs/run.cgi/freqs?q=aword,[lc%20=%22fudbalu%22%20&%20tag=%22N.*%22]&corpname=SrFudKo&refs=doc&fcrit=lemma/e%200~0%3E0&ml=1>;
       #    frac:corpus :BNC;
       #    frac:queryLanguage <https://www.sketchengine.eu/documentation/corpus-querying/>;
       #    frac:result :thes_BNC_le_risk_n.
    
       # collocations
       gq_coll = URIRef(self.url+"q_coll_"+ self.le_url) #query
       gr_coll = URIRef(self.url+"r_coll_"+ self.le_url) #queryResults
        
       g.add( (gq_coll, RDF.type, FRAC.Query))
       g.add( (gq_coll, RDF.value, Literal(self.coll)))
       #g.add( (gq_coll, OWL.sameAs, Literal(self.colli))) # excluded as too long
       g.add( (gq_coll, FRAC.corpus, gcor))
       g.add( (gq_coll, FRAC.queryLanguage, Literal(self.cql_lng_url)))
       g.add( (gq_coll, FRAC.result, gr_coll)) 
       g.add( (gr_coll, RDF.type, FRAC.Cooccurrence))
       g.add( (gr_coll, RDF.type, RDF.Seq))
        
       for item in dColl["Items"][:3]:
          gcolocate = URIRef(self.url+"coll_"+ self.le_url+"_"+item['str'])
          #gcolocate = [URIRef(self.url+"_"+self.le_url), URIRef(self.url+"_"+item['str'])]  
          Seq(g, gr_coll , [gcolocate])
          g.add( (gcolocate, RDF.type, FRAC.Collocation))
          g.add( (gcolocate, RDF.value, Literal(item['freq'])))  
          g.add( (gcolocate, FRAC.obsevedIn, gcor))
        
          # a ontolex:LexicalEntry ; ontolex:lexicalForm [ ontolex:writtenRep "southwestern" ] ;
          g.add( (gcolocate, FRAC.obsevedIn, gcor))
          g.add((gcolocate, DCT.description, Literal(item['str']) ))
          g.add( (gcolocate, FRAC.tScore ,Literal(self.getStats(item['Stats'],'t'))))  
          g.add( (gcolocate, FRAC.logDice,Literal(self.getStats(item['Stats'],'d')))) 
          g.add( (gcolocate, FRAC.pmi,Literal(self.getStats(item['Stats'],'m')))) 
         
          #g.add( (gr_coll, FRAC.collocation, gcolocate))   #a frac:Observable;
        #  g.add( (gcolocate, RDF.type, RDF.Seq))
          #  Seq(g, BNode(), [Literal(self.lemma),  Literal(item['str'])])
          #g.add( (gr_coll,  
       #  gcolocate = URIRef(self.url+"le_"+item['str'])   
        # g.add( (gle,RDF.type,ONTOLEX.LexicalEntry) )
        # g.add( (gle,ONTOLEX.writtenRep, Literal(item['str'] )))
        # g.add( (gle,RDF.type,ONTOLEX.LexicalEntry) ) 
        
        
       # thesaurus             
       gq_thes = URIRef(self.url+"q_thes_"+ self.le_url) #query
       gr_thes = URIRef(self.url+"r_thes_"+ self.le_url) #queryResults
        
       g.add( (gq_thes, RDF.type, FRAC.Query))
       g.add( (gq_thes, RDF.value, Literal(self.thes)))
       g.add( (gq_thes, OWL.sameAs, Literal(self.thesi)))
       g.add( (gq_thes, FRAC.corpus, gcor))
       g.add( (gq_thes, FRAC.queryLanguage, Literal(self.cql_lng_url)))
       g.add( (gq_thes, FRAC.result, gr_thes))
    
       for word in dThes["Words"]:
         gle = URIRef(self.url+"le_"+word['word'])   
     #    g.add( (gle,RDF.type,ONTOLEX.LexicalEntry) )
     #    g.add( (gle,ONTOLEX.writtenRep, Literal(word['word'] )))
     #    g.add( (gle,RDF.type,ONTOLEX.LexicalEntry) ) 

      # WordSketch
       gq_ws = URIRef(self.url+"q_ws_"+ self.le_url) #query
       gr_ws = URIRef(self.url+"r_ws_"+ self.le_url) #queryResults
       
       g.add( (gq_ws, RDF.type, FRAC.Query))
       g.add( (gq_ws, RDF.value, Literal(self.wsketch)))
       g.add( (gq_ws, OWL.sameAs, Literal(self.wsketchi)))
       g.add( (gq_ws, FRAC.corpus, gcor))
       g.add( (gq_ws, FRAC.queryLanguage, Literal(self.cql_lng_url)))
       g.add( (gq_ws, FRAC.result, gr_ws))
        
       for g in dWSketch['Gramrels'][:2]:
        print ('    ' + g['name'])
        for i in g['Words'][:2]:
            print ('        ' + i['word'])
            print(i) 
    
    
        
      # if self.publisher !="" :
       #  g.add( (gnovel, DCT.publisher, Literal(wiki2entity(self.publisher),datatype=XSD.string)) )
        
       return glex
```
